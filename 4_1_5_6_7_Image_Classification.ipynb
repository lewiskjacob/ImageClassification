```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from skimage.transform import resize
from skimage.io import imread
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, ConfusionMatrixDisplay
```

```python
# let's assume data_array as the inputs (x) and target_array as the output (y). Whenever we add a cat image to data_arrray, we add 0 to the target array and
# for dog, we will add 1 in data_array
```

```python
data_array=[]
target_array=[]
categories=['Cat','Dog']
datadir='/content/drive/MyDrive/Data'
for i in categories:
  sub_path=os.path.join(datadir,i)
  images=os.listdir(sub_path)
  for img in images:
    img_path=os.path.join(sub_path,img) # here we got path of all the pics in cat and dog folder
    img_array=imread(img_path)
    img_resized=resize(img_array,(150,150,3))
    data_array.append(img_resized.flatten())
    target_array.append(categories.index(i))
  print("Loaded category:",i)
```

```python
data=np.array(data_array)
data.shape
```

```python
target=np.array(target_array)
target.shape
```

```python
df=pd.DataFrame(data)
df # these are the pixel values of RGB,, but it is btwn 0 to 1 instead of 0 to 255. ie,these are scaled when we did the resizing of the image
```

```python
df['target']=target
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(data,target,test_size=0.3,random_state=1)
```

```python
sv=SVC()
sv.fit(x_train,y_train)
y_pred=sv.predict(x_test)
y_pred
```

```python
print(classification_report(y_test,y_pred))
print(ConfusionMatrixDisplay.from_predictions(y_test,y_pred))
```
